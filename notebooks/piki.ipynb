```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
earnings = pd.read_csv('data/coins-come.csv', 
                       parse_dates=['Дата'],
                       sep=';',
                       encoding='utf-8-sig',
                       dayfirst=False)
purchases = pd.read_csv('data/coins-out.csv', 
                        parse_dates=['Дата оформления заказа'], 
                        sep=';',
                        encoding='utf-8-sig',
                        dayfirst=False)
employees = pd.read_excel('data/personal-list.xlsx')
```

```python
purchases['Дата оформления заказа'] = pd.to_datetime(
    purchases['Дата оформления заказа'], 
    format='%Y-%m-%d %H:%M:%S')
```

```python
purchases['Месяц'] = purchases['Дата оформления заказа']
```

```python
import locale
locale.setlocale(locale.LC_TIME, 'ru_RU.UTF-8')
```

```python
purchases['Месяц'] = purchases['Дата оформления заказа'].dt.strftime('%B %Y')  
purchases['Месяц_число'] = purchases['Дата оформления заказа'].dt.month 
```

```python
activity = (
    purchases.groupby(['Месяц', 'Статус товара'])
    .agg(Покупки=('Наименование товара', 'count'))
    .reset_index()
)

month_order = purchases.sort_values('Месяц_число')['Месяц'].unique()
activity['Месяц'] = pd.Categorical(activity['Месяц'], categories=month_order, ordered=True)
```

```python
plt.figure(figsize=(16, 8))
sns.lineplot(
    data=activity,
    x='Месяц',
    y='Покупки',
    hue='Статус товара',
    style='Статус товара',
    markers=True,
    dashes=False,
    palette='tab10',
    linewidth=2.5,
    markersize=10
)

plt.title('Динамика покупок по статусам товаров (с учетом времени)', fontsize=16)
plt.xlabel('Месяц', fontsize=12)
plt.ylabel('Количество покупок', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Статус товара', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 1. Удаление выброса
purchases_clean = purchases[~purchases['Наименование товара'].str.contains('Лотерейный билет "Надо собраться"', na=False)]

# 2. Фильтрация топ-12 товаров (после удаления выброса)
top_items = purchases_clean['Наименование товара'].value_counts().nlargest(12).index
purchases_filtered = purchases_clean[purchases_clean['Наименование товара'].isin(top_items)]

# 3. Создание временных меток
purchases_filtered['Год'] = purchases_filtered['Дата оформления заказа'].dt.year
purchases_filtered['Месяц'] = purchases_filtered['Дата оформления заказа'].dt.month
purchases_filtered['Дата_период'] = purchases_filtered['Дата оформления заказа'].dt.to_period('M')

# 4. Группировка данных
activity = (
    purchases_filtered.groupby(['Дата_период', 'Год', 'Месяц', 'Наименование товара'])
    .agg(Покупки=('Наименование товара', 'count'))
    .reset_index()
    .sort_values(['Год', 'Месяц'])
)

# 5. Форматирование для визуализации
activity['Период'] = activity['Дата_период'].dt.strftime('%b %Y')
activity['Порядок'] = activity['Год'] * 100 + activity['Месяц']

# 6. Визуализация с увеличенным масштабом
plt.figure(figsize=(25, 12))
ax = sns.lineplot(
    data=activity,
    x='Период',
    y='Покупки',
    hue='Наименование товара',
    palette='husl',  # Более контрастная палитра
    linewidth=2.5,
    marker='D',
    markersize=10,
    style='Наименование товара',
    dashes=False,
    sort=False
)

# 7. Настройка отображения
ax.set_xticks(range(len(activity['Период'].unique())))
ax.set_xticklabels(
    activity.drop_duplicates('Порядок')['Период'].values,
    rotation=45,
    ha='right',
    fontsize=12
)

# 8. Оптимизация осей
max_purchases = activity['Покупки'].max()
plt.ylim(0, max_purchases * 1.1)  # Добавление 10% запаса сверху
plt.yticks(fontsize=12)

# 9. Легенда и заголовки
plt.title('Динамика покупок по топ-12 товарам (без выбросов)', fontsize=18, pad=20)
plt.xlabel('Период', fontsize=14)
plt.ylabel('Количество покупок', fontsize=14)
plt.legend(
    title='Товары',
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    fontsize=12,
    title_fontsize=14
)

# 10. Сетка и оформление
plt.grid(alpha=0.4, linestyle='--')
plt.tight_layout()

plt.ylim(0, 220)  # Максимальное значение по вашему набору данных

# 11. Анализ результатов
print(f"Максимальное количество покупок после фильтрации: {max_purchases}")
plt.show()
```
