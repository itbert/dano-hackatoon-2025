```python
import matplotlib.pyplot as plt
import pandas as pd

# Данные
data = {
    "Товар": [
        "Лотерея",
        "Кофе",
        "Внешний аккумулятор Xiaomi",
        "Подписка Яндекс плюс",
        "Худи КРОК Hard",
        "Футболка КРОК Hard",
        "Носки Soft&Hard",
        "Худи КРОК",
        "Футболка КРОК Soft",
        "Футболка \"Надо собраться\""
    ],
    "Количество": [
        6733,
        5649,
        785,
        781,
        582,
        548,
        511,
        484,
        444,
        387
    ]
}

# Создаем DataFrame
df = pd.DataFrame(data)

# Создаем график
plt.figure(figsize=(10, 6))  # Устанавливаем размер графика
plt.bar(df["Товар"], df["Количество"], color="#87CEEB")  # Создаем столбчатую диаграмму с синим цветом
plt.title("Распределение товаров по количеству", fontsize=16)  # Устанавливаем заголовок с размером шрифта
plt.xlabel("Товар", fontsize=14)  # Подписываем ось X с размером шрифта
plt.ylabel("Количество", fontsize=14)  # Подписываем ось Y с размером шрифта
plt.xticks(rotation=45, ha='right')  # Поворачиваем метки по оси X для лучшей читаемости
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Добавляем сетку по оси Y

# Настройка внешнего вида
plt.tight_layout()  # Убираем обрезанные метки
plt.style.use('seaborn')  # Используем стиль seaborn для более привлекательного вида

# Показываем график
plt.show()
```
