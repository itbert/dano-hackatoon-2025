```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
%matplotlib inline
```

```python

# 1. Загрузка данных
df = pd.read_excel('data/personal-list.xlsx')

# 2. Первичный осмотр данных
print("Первые 5 строк:")
display(df.head())

print("\nИнформация о данных:")
df.info()

print("\nРазмер данных:", df.shape)
print("\nКолонки:", df.columns.tolist())

# 3. Анализ пропущенных значений
print("\nПропущенные значения:")
missing = df.isnull().sum()
print(missing[missing > 0].sort_values(ascending=False))

# Визуализация пропущенных значений (опционально)
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False)
plt.title('Распределение пропущенных значений')
plt.show()

# 4. Проверка дубликатов
print("\nКоличество дубликатов:", df.duplicated().sum())

# 5. Статистический анализ
print("\nОписательная статистика для числовых признаков:")
display(df.describe().T)

print("\nАнализ категориальных признаков:")
categorical = df.select_dtypes(include=['object'])
for col in categorical:
    print(f"\n{col}:")
    display(categorical[col].value_counts(normalize=True))

# 6. Визуализация распределений
# Для числовых признаков
numerical = df.select_dtypes(include=['int64', 'float64'])
for col in numerical:
    plt.figure(figsize=(10,4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Распределение {col}')
    plt.show()

# Для категориальных признаков
for col in categorical:
    plt.figure(figsize=(10,4))
    sns.countplot(data=df, x=col)
    plt.title(f'Распределение {col}')
    plt.xticks(rotation=45)
    plt.show()

# 7. Анализ выбросов
plt.figure(figsize=(12,6))
sns.boxplot(data=numerical)
plt.title('Распределение числовых признаков (масштабировано)')
plt.xticks(rotation=45)
plt.show()

# 8. Корреляционный анализ
corr_matrix = numerical.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Матрица корреляций')
plt.show()

# 9. Анализ целевой переменной (если есть)
# Пример для задачи классификации
target = 'your_target_column'
if target in df:
    plt.figure(figsize=(10,5))
    sns.countplot(data=df, x=target)
    plt.title('Распределение целевой переменной')
    plt.show()

# 10. Предварительные выводы
print("\nПредварительные выводы:")
print("-"*50)
print(f"Всего наблюдений: {df.shape[0]}")
print(f"Количество признаков: {df.shape[1]}")
print(f"Пропущенные значения: {df.isnull().sum().sum()} ({round(df.isnull().mean().mean()*100, 2)}%)")
print(f"Дубликаты: {df.duplicated().sum()}")
print("Рекомендации по следующим шагам:")
print("1. Обработка пропущенных значений")
print("2. Обработка выбросов")
print("3. Кодирование категориальных признаков")
print("4. Feature engineering")
print("5. Нормализация/стандартизация данных при необходимости")
```
