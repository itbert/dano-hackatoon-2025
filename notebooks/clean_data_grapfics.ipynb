```python
import pandas as pd 
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
import re
```

```python
spisok = pd.read_excel("data/personal-list.xlsx")
zarabotok = pd.read_csv("data/coins-come.csv", sep=";")
pokupki = pd.read_csv("data/coins-out.csv", sep=";")
```

```python
def months(staj):
    match = re.findall(r'(\d+)', staj)
    if len(match) == 3:
        years, months, days = map(int, match)
        total_months = years * 12 + months + (days // 30)
        return total_months
    return None

spisok['months_total'] = spisok["Стаж фактический по компании"].apply(months)

zarabotok['Fact_dep'] = zarabotok['Департамент'].str.split('.').str[0]
zarabotok
```

```python
gender_dict = dict(zip(spisok['Внешний код'], spisok['Пол']))

# Добавляем столбец с полом в первый датафрейм
zarabotok['gender'] = zarabotok['Код сотрудника'].map(gender_dict)
```

```python
gender_dict = dict(zip(zarabotok['Код сотрудника'], zarabotok.groupby(["Код сотрудника"])['Сумма вознаграждения'].sum()))

# Добавляем столбец с полом в первый датафрейм
spisok['zarab_bobor'] = spisok['Внешний код'].map(gender_dict)


spisok

sns.boxplot(y=spisok["zarab_bobor"])


df = spisok[spisok["zarab_bobor"] > 5000]

res = spisok.groupby("Пол")["zarab_bobor"].median()

spisok["zarab_bobor"].quantile(0.75)
spisok["zarab_bobor"].quantile(0.25)

# 4213 это 75 квартиль плюс 1.5 меж квартильных размаха
spisok
```

```python
user_totals = spisok.groupby('Внешний код')['zarab_bobor'].sum().reset_index()

user_totals

users_to_remove = user_totals[user_totals['zarab_bobor'] >= 4212]['Внешний код']

# # Шаг 3: Удаляем все заказы этих пользователей из исходного DataFrame
pokupki = pokupki[~pokupki['Код сотрудника'].isin(users_to_remove)]

pokupki
```

```python

def lot(str):
    if str.startswith("Лоте") or str.startswith("Черный"):
        return "Лотерея"
    elif str.startswith("Кофе"):
        return "Кофе"
    else:
        return str


pokupki['tovar'] = pokupki["Наименование товара"].apply(lot)
index_to_drop = pokupki[(pokupki["Стоимость в валюте"] == 0) & (pokupki["Скидка"] != 100)].index
pokupki = pokupki.drop(index_to_drop)

pokupki["tovar"].value_counts()
```

```python
plt.figure(figsize=(10, 6))

data = user_totals['zarab_bobor']

threshold = 4212
mask_removed = data >= threshold
mask_kept = data < threshold

bins = np.linspace(data.min(), data.max(), 30)
bin_width = bins[1] - bins[0]

counts_kept, _ = np.histogram(data[mask_kept], bins=bins)
counts_removed, _ = np.histogram(data[mask_removed], bins=bins)

plt.barh(bins[:-1], counts_kept, height=bin_width, 
         color='green', alpha=0.7, 
         label='Оставили')
plt.barh(bins[:-1], counts_removed, height=bin_width, 
         color='red', alpha=0.7, 
         label='Удалили')

plt.axhline(threshold, color='black', 
            linestyle='--', 
            label=f'Порог: {threshold}')

plt.title('Распределение суммы вознаграждения пользователей')
plt.xlabel('Количество пользователей')
plt.ylabel('Сумма вознаграждения')   
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(0, 200)  

plt.show()
```

```python
plt.figure(figsize=(12, 8))
df_plot = user_totals[['zarab_bobor']].copy()
threshold = 4212
df_plot['status'] = np.where(df_plot['zarab_bobor'] < threshold, 
                            'Оставшиеся', 'Удаленные')

colors = {'Оставшиеся': '#2ecc71', 'Удаленные': '#e74c3c'}
plt.rcParams['font.family'] = 'Arial'

ax = sns.histplot(
    data=df_plot,
    x='zarab_bobor',
    hue='status',
    palette=colors,
    bins=30,
    kde=True,
    edgecolor='#34495e',
    linewidth=0.5,
    element='bars',
    multiple='stack'
)

total_kept = len(df_plot[df_plot['status'] == 'Оставшиеся'])
total_removed = len(df_plot) - total_kept

ax.text(0.95, 0.95, 
       f"Сохранилось: {total_kept}\nУдалено: {total_removed}",
       transform=ax.transAxes,
       verticalalignment='top',
       horizontalalignment='right',
       bbox=dict(facecolor='white', alpha=0.8))
plt.axvline(threshold, color='#2c3e50', ls='--', lw=2)
plt.title('Распределение пользовательских вознаграждений\n', fontsize=14)
plt.xlabel('Сумма вознаграждения', fontsize=12)
plt.ylabel('Количество пользователей', fontsize=12)
plt.grid(axis='y', linestyle=':', alpha=0.7)
sns.despine(left=True)

# plt.legend(title='Категория', 
#           labels=[f'Оставшиеся (n={total_kept})', 
#                   f'Удаленные (n={total_removed})'])
plt.tight_layout()

plt.ylim(0, 300)
plt.xlim(0, 10000)

plt.show()
```
